```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit


# Dfinir la fonction de l'exponentielle tire
def stretched_exp(x, A, tau, beta, offset):
    return A * np.exp(-(x/tau)**beta)+offset

# Ajuster les paramtres
popt, pcov = curve_fit(stretched_exp, x, y.real, p0=[1e5, 200e3, 1.5,1e5])

# Extraire les paramtres ajusts
A, tau, beta, offset = popt

# Tracer les rsultats
plt.scatter(x, y, label='Donnes')
plt.plot(x, stretched_exp(x, *popt), 'r-', label='Ajustement')
plt.legend()
plt.show()

print(f"Paramtres ajusts: A={A}, tau={tau}, beta={beta}")
```
