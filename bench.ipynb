```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit


# Définir la fonction de l'exponentielle étirée
def stretched_exp(x, A, tau, beta, offset):
    return A * np.exp(-(x/tau)**beta)+offset

# Ajuster les paramètres
popt, pcov = curve_fit(stretched_exp, x, y.real, p0=[1e5, 200e3, 1.5,1e5])

# Extraire les paramètres ajustés
A, tau, beta, offset = popt

# Tracer les résultats
plt.scatter(x, y, label='Données')
plt.plot(x, stretched_exp(x, *popt), 'r-', label='Ajustement')
plt.legend()
plt.show()

print(f"Paramètres ajustés: A={A}, tau={tau}, beta={beta}")
```
